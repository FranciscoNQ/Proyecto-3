## **Introducción**

Este análisis revela las diferencias de poder adquisitivo entre los países de Sudamérica, basándose en la cantidad de Big Mac que se puede adquirir con el salario mínimo expresado en dólares.
Además, permite observar el costo de vida en el ámbito gastronómico a través del precio de la Big Mac (también conocido como Índice Big Mac), así como el nivel del salario mínimo de cada país en moneda fuerte (USD).

## Inicio: Carga del dataset

```python
#Libreria
import pandas as pd

#Variable con ubicacion y nombre del dataset
ubiacion = "/content/drive/MyDrive/Proyecto #3/"
dataset = "dataset_final"

#Incio del dataframe
df = pd.read_csv(ubiacion + dataset)
```

## **Informe: Salario mínimo, precio de Big Mac y poder adquisitivo**

```python
df.sort_values(by="Salario minimo (USD)", ascending=False, inplace=True)
salario_minimo_mas_alto = df.head(1)
salario_minimo_mas_bajo = df.tail(2)
print("Salario Minimo (USD):")
print("* Mayor:", salario_minimo_mas_alto["Pais"].iloc[0]+ " - " + str(salario_minimo_mas_alto["Salario minimo (USD)"].iloc[0]), "USD")
print("* Menor:", salario_minimo_mas_bajo["Pais"].iloc[0]+ " - " + str(salario_minimo_mas_bajo["Salario minimo (USD)"].iloc[0]), "USD")
print("* Promedio:", df["Salario minimo (USD)"].mean().round(2), "USD")
```

image.png

```python
df.sort_values(by="Precio bigmac (USD)", ascending=False, inplace=True)
precio_bigmac_mas_alto = df.head(1)
precio_bigmac_mas_bajo = df.tail(1)
print("Precio Big Mac (USD):")
print("* Mayor:", precio_bigmac_mas_alto["Pais"].iloc[0]+" - "+str(precio_bigmac_mas_alto["Precio bigmac (USD)"].iloc[0]), "USD")
print("* Menor:", precio_bigmac_mas_bajo["Pais"].iloc[0]+" - "+str(precio_bigmac_mas_bajo["Precio bigmac (USD)"].iloc[0]), "USD")
print("* Promedio:", df["Precio bigmac (USD)"].mean().round(2), "USD")
```

image.png

```python
df.sort_values(by="Cantidad_bigmac_con_salario", ascending=False, inplace=True)
cantidad_bigmac_mayor_x_salario = df.head(1)
cantidad_bigmac_menor_x_salario = df.tail(2)
print("Poder Adquisitivo:")
print("* Mayor:", cantidad_bigmac_mayor_x_salario["Pais"].iloc[0]+" - "+str(cantidad_bigmac_mayor_x_salario["Cantidad_bigmac_con_salario"].iloc[0]), "Big mac")
print("* Menor:", cantidad_bigmac_menor_x_salario["Pais"].iloc[0]+" - "+str(cantidad_bigmac_menor_x_salario["Cantidad_bigmac_con_salario"].iloc[0]), "Big Mac")
print("* Promedio:", df["Cantidad_bigmac_con_salario"].mean().round(2), "Big Mac")
```

## **Conclusión**

Se observa que, aunque un país tenga un salario mínimo alto, si el costo de vida también es elevado, el poder adquisitivo puede ser considerablemente menor. Un ejemplo de esto es Uruguay.
También se puede dar el caso contrario, como en Brasil: tiene el salario más bajo de la comparación, pero también presenta el costo de vida más bajo, lo que da como resultado un poder adquisitivo por encima del promedio.
El peor escenario se da cuando el salario mínimo es bajo y el costo de vida alto, lo que genera una fuerte pérdida del poder adquisitivo. Un claro ejemplo de esto es Argentina, que se posiciona como el país con peor poder adquisitivo de la región.



image.png

En el gráfico se muestra el nivel de poder adquisitivo, expresado en porcentaje.
Al comparar el país con mayor poder adquisitivo (Chile) con los últimos, la diferencia es cercana al doble. Pero, si la comparación se hace con el país con menor poder adquisitivo (Argentina), la brecha se amplía considerablemente, siendo casi el triple.
